```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("delhiaqi.csv")
```

```python
print(df.head())

print(df.columns)

df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['date'], df['pm2_5'])

plt.title("Delhi PM2.5 Trend Over Time")
plt.xlabel("Date")
plt.ylabel("PM2.5")

plt.xticks(rotation=45)

plt.show()
```

```python
monthly_pm = df.groupby('month')['pm2_5'].mean()

plt.figure(figsize=(10,5))

monthly_pm.plot(kind='bar')

plt.title("Monthly Average PM2.5")
plt.xlabel("Month")
plt.ylabel("Average PM2.5")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['pm2_5'], bins=30)

plt.title("PM2.5 Distribution")
plt.xlabel("PM2.5")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    df[['co','no','no2','o3','so2','pm2_5','pm10','nh3']].corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(x='month', y='pm2_5', data=df)

plt.title("Monthly PM2.5 Variation")
plt.xlabel("Month")
plt.ylabel("PM2.5")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='pm10', y='pm2_5', data=df)

plt.title("PM10 vs PM2.5")
plt.xlabel("PM10")
plt.ylabel("PM2.5")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(df['pm2_5'], fill=True)

plt.title("PM2.5 Density Plot")
plt.xlabel("PM2.5")

plt.show()
```

```python
yearly_pm = df.groupby('year')['pm2_5'].mean()

plt.figure(figsize=(8,5))

yearly_pm.plot(marker='o')

plt.title("Year-wise PM2.5 Trend")
plt.xlabel("Year")
plt.ylabel("Average PM2.5")

plt.show()
```
