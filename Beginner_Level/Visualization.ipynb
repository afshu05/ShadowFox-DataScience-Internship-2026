```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
days = [1, 2, 3, 4, 5]
sales = [15, 30, 25, 40, 60]

plt.plot(days, sales)

plt.title("Line Chart - Daily Sales Trend ")
plt.xlabel("Day Number")
plt.ylabel("Sales Amount")

plt.show()
```

```python
marks = [35, 40, 42, 45, 47, 50, 52, 55, 58, 60, 62, 65, 67, 70, 72, 75, 78, 80, 82, 85, 87, 88, 90, 91, 92, 93, 94, 95, 96, 97]

plt.hist(marks, bins=6, color='steelblue')

plt.title("Student Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")

plt.show()
```

```python
cities = ['Pune', 'Mumbai', 'Delhi', 'Pune', 'Mumbai', 'Delhi', 'Chennai', 'Chennai']
temperature = [30, 32, 35, 31, 33, 36, 29, 30]

plt.scatter(cities, temperature)

plt.title("Scatter Plot - City Temperature Analysis")
plt.xlabel("Cities")
plt.ylabel("Temperature (°C)")

plt.show()
```

```python
fruits = ['Apple', 'Banana', 'Mango', 'Orange', 'Grapes']
sales = [30, 20, 25, 15, 10]

plt.pie(sales, labels=fruits, autopct='%1.1f%%')

plt.title("Pie chart - Fruit Sales Distribution")

plt.show()
```

```python

# SEABORN PLOTS


# Load datasets used by Seaborn examples
tips = sns.load_dataset("tips")
penguins = sns.load_dataset("penguins")
df = penguins.dropna()   # used for histplot & heatmap


# 6. SEABORN LINE PLOT ──
import seaborn as sns
import matplotlib.pyplot as plt

sns.lineplot(data=tips, x="size", y="tip")

plt.show()
```

```python
sns.load_dataset("tips")

sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker")

plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill")

plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill")

plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill")

plt.show()
```

```python
sns.displot(data=tips, x="total_bill", kind="kde")

plt.show()
```

```python
sns.histplot(x='bill_length_mm', data=df, palette="Set2")
plt.title('Histogram Plot ')
plt.ylabel('flipper length')
plt.xlabel('Species')
plt.show()
```

```python
penguins_encoded = pd.get_dummies(df, columns=['species', 'island', 'sex'], drop_first=True)

correlation_matrix = penguins_encoded.corr()

plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Penguin Data")
plt.show()
```

```python
sns.regplot(x="bill_length_mm", y="flipper_length_mm", data=penguins)
plt.title("Regression Plot of Bill Length vs. Flipper Length")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```
